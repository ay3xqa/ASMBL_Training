```python
import shutil
```

```python
path = Path('gdrive/MyDrive/ASMBL_Training')
path.ls(file_exts='.pkl')
```

```python
learn_inf = load_learner(path/'asmbl_model.pkl')
learn_inf.dls.vocab
```

```python
files = get_image_files("gdrive/MyDrive/ASMBL_Training/nvidia+geforce+rtx+3060+ti+founders+edition")
valid = 0
for index in range(len(files)):
  img = PILImage.create(files[index])
  pred,pred_idx,probs = learn_inf.predict(img)
  dest = str(files[index]).replace("nvidia+geforce+rtx+3060+ti+founders+edition","rtx3060-cleaned")
  if pred == "Valid":
    valid+=1
    shutil.copyfile(files[index], dest)
print("Valid items: "+str(valid))
```
